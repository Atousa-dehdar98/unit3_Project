```python
# List all columns with missing values

missing_values = data.isnull().sum()

missing_values = missing_values[missing_values > 0]

missing_values
```

## Missing Value Analysis

Missing values can negatively affect machine learning algorithms.

The above output shows all columns containing missing values and the number of missing records in each column.

Since removing rows may lead to information loss, missing values will be imputed instead.

```python
# Fill missing values in ALB

data['ALB'].fillna(data['ALB'].median(), inplace=True)

print(data['ALB'].isnull().sum())
```

### ALB

The ALB feature contains numerical laboratory measurements.

Median imputation is used because it is robust against outliers.

The output confirms that no missing values remain in this column.

```python
data['ALP'].fillna(data['ALP'].median(), inplace=True)

print(data['ALP'].isnull().sum())
```

### ALP

Missing values in ALP were replaced using the median value of the column.

Median imputation helps preserve the overall distribution of the data.

```python
data['ALT'].fillna(data['ALT'].median(), inplace=True)

print(data['ALT'].isnull().sum())
```

### ALT

Missing ALT values were imputed using the median.

This avoids removing observations from the dataset.

```python
data['CHOL'].fillna(data['CHOL'].median(), inplace=True)

print(data['CHOL'].isnull().sum())
```

### CHOL

Missing cholesterol measurements were filled using median imputation.

The median is preferred because laboratory values may contain extreme observations.

```python
data['PROT'].fillna(data['PROT'].median(), inplace=True)

print(data['PROT'].isnull().sum())
```

### PROT

Missing protein measurements were replaced by the median value.

No rows were removed during the cleaning process.

```python
# Verify no missing values remain

data.isnull().sum()
```

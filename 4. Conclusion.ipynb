```python
# Summary of actual classes within each predicted class

cluster_summary = pd.crosstab(
    df['Predicted_class'],
    df['Category'],
    normalize='index'
) * 100

cluster_summary = cluster_summary.round(2)

print("Percentage distribution of actual classes within each predicted class:")
display(cluster_summary)
```

## 3. Explanations and isights :

```python
# in this section you should display analysis about each cluster and comment those one by one 

print("Cluster / Class Analysis")
print("="*60)

for predicted_class in cluster_summary.index:

    print(f"\nPredicted Class: {predicted_class}")

    dominant_class = cluster_summary.loc[predicted_class].idxmax()
    dominant_percentage = cluster_summary.loc[predicted_class].max()

    print(
        f"Dominant actual class: {dominant_class} "
        f"({dominant_percentage:.2f}%)"
    )

    print("\nClass distribution:")

    for actual_class, percentage in cluster_summary.loc[predicted_class].items():

        if percentage > 0:
            print(f"  {actual_class}: {percentage:.2f}%")
```

### Interpretation of Results

The predicted Blood Donor class is highly pure and contains the majority of
healthy patients. This indicates that the model can successfully recognize
normal liver profiles.

The Hepatitis and Fibrosis classes show some overlap. This behaviour is
expected because both diseases share similar biochemical characteristics,
making them difficult to separate perfectly.

The Cirrhosis class is identified with high precision. Most samples predicted
as Cirrhosis correspond to actual Cirrhosis patients, which demonstrates that
the model can capture the severe liver damage patterns associated with this
condition.

Overall, the model performs best on the Blood Donor and Cirrhosis classes,
while Hepatitis and Fibrosis remain the most challenging categories.
